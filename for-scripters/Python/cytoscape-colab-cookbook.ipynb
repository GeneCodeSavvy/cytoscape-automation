## **Open session file from cloud drive**

To load a file from cloud storage, use the file's URL and the ``sandbox_url_to`` function to download

```python
p4c.sandbox_url_to('https://drive.google.com/file/d/1ZFifBC8N6k5EU5Gh35VJZwDBr7k6J0Kw/view?usp=sharing','mySession.cys','default_sandbox')
p4c.open_session('mySession.cys')
```

## **Color palette mapping generators**

The palette functions help generate color gradient mappings based on Table data, all in a single line of code!

```python
p4c.set_node_color_mapping(**p4c.gen_node_color_map('gal80Rexp', p4c.palette_color_brewer_d_RdBu(), style_name='galFiltered Style'))
```

## **Inline display of network PNGs**

py4cytoscape 0.0.11 add notebook_export_show_image() and notebook_show_image() functions. RCy3 will also have these functions.

```python
p4c.notebook_export_show_image(filename='my_image.png')
```
